### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
!pip install geopy
```

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score
# Print the results
# Find the coordinates of "Penang Flavours"
penang_flavours_location = (51.49014200, 0.08384000)  # Latitude and Longitude

# Define the maximum allowable difference in latitude and longitude in degrees
max_latitude_difference = 0.01
max_longitude_difference = 0.01

# Query the database to find establishments with RatingValue = 5
results = db[collection_name].find(
    {
        "RatingValue": 5,
        "geocode.latitude": {"$exists": True},
        "geocode.longitude": {"$exists": True}
    }
)

# Create a list to store the results within the specified latitude and longitude range
establishments_within_range = []

for establishment in results:
    establishment_location = (float(establishment["geocode"]["latitude"]), float(establishment["geocode"]["longitude"]))
    latitude_difference = abs(penang_flavours_location[0] - establishment_location[0])
    longitude_difference = abs(penang_flavours_location[1] - establishment_location[1])
    
    # Check if the establishment is within the specified range
    if latitude_difference <= max_latitude_difference and longitude_difference <= max_longitude_difference:
        establishments_within_range.append(establishment)

# Sort the establishments by hygiene score
top_5_establishments = sorted(establishments_within_range, key=lambda x: x["scores"]["Hygiene"])[:5]

# Print the top 5 establishments
for idx, establishment in enumerate(top_5_establishments, start=1):
    print(f"Top {idx} Establishment:")
    pprint(establishment)
```

```python
# Convert result to Pandas DataFrame
results_list = list(top_5_establishments)

df = pd.DataFrame(results_list)

print(df)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
hygiene_score = 0
# 2. Groups the matches by Local Authority
pipeline = [
    {
        "$match": {
            "scores.Hygiene": hygiene_score
        }
    },
    {
        "$group": {
            "_id": "$LocalAuthorityName",
            "count": {"$sum": 1}
        }
    },
    {
        "$sort": {"count": -1}
    }
]

result = list(db[collection_name].aggregate(pipeline))
# 3. Sorts the matches from highest to lowest
# Print the number of documents in the result
# Print the first 10 results
# Print the number of documents in the result
print("Number of establishments with a hygiene score of 0 in each Local Authority area:")
for item in result:
    print(f"{item['_id']}: {item['count']} establishments")

# Print the first 10 results
print("\nFirst 10 results:")
pprint(result[:10])
```

```python
result = list(db[collection_name].aggregate(pipeline))
# Convert the result to a Pandas DataFrame
df = pd.DataFrame(result)

# Display the number of rows in the DataFrame
num_rows = len(df)
print(f"Number of rows in the DataFrame: {num_rows}")

# Display the first 10 rows of the DataFrame
print("\nFirst 10 rows of the DataFrame:")
print(df.head(10))
```
